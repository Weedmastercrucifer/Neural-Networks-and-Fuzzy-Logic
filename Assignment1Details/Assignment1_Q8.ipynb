```python
df=data.iloc[:][:].values
np.random.shuffle(df)
```

```python
X=df[:,0:30]
y=df[:,30:31]
for i in range(569):
  if(y[i]=='M'):
    y[i]=1
  else:
    y[i]=0
```

```python
a=X.mean(axis=0)
b=X.max(axis=0)-X.min(axis=0)
Xnorm=np.zeros((X.shape[0],X.shape[1]))
for i in range(30):
  Xnorm[:,i:i+1]=(X[:,i:i+1]-a[i])/b[i]
```

```python
def value(X,y):
  sum=np.zeros((2,X.shape[1]))
  count=0
  count1=0
  X0=X.copy()
  X1=X.copy()
  for i in range(y.shape[0]):
    if (y[i]==0):
      sum[0]=sum[0]+X[i]
      X1=np.delete(X1,i-count,0)
      count=count+1
    else:
      sum[1]=sum[1]+X[i]
      X0=np.delete(X0,i-count1,0)
      count1=count1+1
  mean=np.zeros((2,X.shape[1]))
  mean[0]=sum[0]/count
  mean[1]=sum[1]/(y.shape[0]-count)
  ratio=count1/count
  cov0=np.cov(X0.T)
  cov1=np.cov(X1.T)
  return mean,ratio,cov0,cov1
```

```python
def LRT(X,mean,ratio,cov0,cov1):
  y_pred=2*np.ones((X.shape[0],1))
  b0=np.linalg.det(cov0)**0.5
  b1=np.linalg.det(cov1)**0.5
  for i in range(X.shape[0]):
    mat0=((X[i]-mean[0])@np.linalg.inv(cov0))@(X[i]-mean[0]).T
    mat1=((X[i]-mean[1])@np.linalg.inv(cov1))@(X[i]-mean[1]).T
    mat0=math.exp(-0.5*mat0)
    mat1=math.exp(-0.5*mat1)
    L0=mat0/b0
    L1=mat1/b1
    if (L0/L1 > ratio):
      y_pred[i]=0
    else:
      y_pred[i]=1
  return y_pred
```

```python
for k in range(5):
  count=0
  if (k!=4):
    X_te=Xnorm[114*k:114*(k+1)]
    y_te=y[114*k:114*(k+1)]
    X_tr=np.delete(Xnorm,slice(114*k,114*(k+1)),0)
    y_tr=np.delete(y,slice(114*k,114*(k+1)),0)
  else:
    X_te=Xnorm[456:]
    y_te=y[456:]
    X_tr=np.delete(Xnorm,slice(456,569),0)
    y_tr=np.delete(y,slice(456,569),0)
  for i in range(y_te.shape[0]):
    if (y_te[i]==1):
      count=count+1
  mean,ratio,cov0,cov1=value(X_tr,y_tr)
  y_pred=LRT(X_te,mean,ratio,cov0,cov1)
  A=y_pred-y_te
  a=np.sum(abs(A))
  acc=100*(len(y_te)-a)/len(y_te) 
  print("Accuracy")
  print(acc)
  b=0.5*(a+np.sum(A))
  c=a-b
  sen=100*(count-c)/count 
  print("Sensitivity")
  print(sen)
  spe=100*(len(y_te)-count-b)/(len(y_te)-count) 
  print("Specificity")
  print(spe)
  print("\n")
```
